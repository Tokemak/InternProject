# Price Elasticity 

##### calculating elasticity of each asset pair 

```python
# set price as apr measurement and smoothen it. 
stETH_ETH_apr_df['cur_naive_apr'] = stETH_ETH_apr_df[['raw_base_apr', 'raw_fee_apr', 'raw_incentive_apr', 'raw_price_apr']].sum(axis=1)
stETH_ETH_apr_df['rolling_cur_naive_apr'] = stETH_ETH_apr_df['cur_naive_apr'].rolling(7).mean()
# price shifted back 30 days. this will be our price_1
stETH_ETH_apr_df['prev_rolling_cur_naive_apr'] = stETH_ETH_apr_df['rolling_cur_naive_apr'].shift(-30)

# set supply (liquidity) for time delay
stETH_ETH_apr_df['prev_token_supply'] = stETH_ETH_apr_df['lp_token_total_supply'].shift(-30)

# set variables for elasticity formula 
quantity_2 = stETH_ETH_apr_df['lp_token_total_supply']
quantity_1 = stETH_ETH_apr_df['prev_token_supply']
price_2 = stETH_ETH_apr_df['rolling_cur_naive_apr']
price_1 = stETH_ETH_apr_df['prev_rolling_cur_naive_apr']

# calculate elasticity 
percent_change_in_supply = (quantity_2 - quantity_1) / ((quantity_2 + quantity_1) / 2) * 100 
percent_change_in_price = (price_2 - price_1) / ((price_2 + price_1) / 2) * 100 
stETH_ETH_apr_df['elasticity'] = percent_change_in_supply / percent_change_in_price
```
